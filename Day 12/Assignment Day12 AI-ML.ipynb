```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("F:\\LetsUpgrade\\Notes\\DAY 7 Notes + Ass\\general_data.csv")
```

```python
dataset.head()
```

```python
dummy=pd.get_dummies(dataset['Attrition'])
dummy
```

```python
dataset2=pd.concat((dataset,dummy),axis=1)
dataset2
```

```python
dataset2=dataset2.drop(['Attrition'],axis=1)
dataset2
```

```python
dataset2 = dataset2.drop(["No"],axis=1)
dataset2
```

```python
dataset2=dataset2.rename(columns={"Yes":"Attrition"})
dataset2.head()
```

# Wilcoxon - Sign Test
### Distance from home and StandardHours

```python
from scipy.stats import wilcoxon

stats,p=wilcoxon(dataset2.DistanceFromHome,dataset2.StandardHours)
print(stats,p)
```

### H0: There is no significant differences in the distance from home and StandardHours
### Ha: There is significant differences in the distance from home and StandardHours
### The p value is 0.043962321781016465 ,So p value is less than 0.5
### Hence Ho (null hypothesis) is rejected and Ha (alternative hypothesis) accepted.
### There is significant differences in the distance from home and StandardHours

# Friedman Test
### YearsAtCompany	and YearsSinceLastPromotion	and YearsWithCurrManage

```python
from scipy.stats import friedmanchisquare

stats,p = friedmanchisquare(dataset2.YearsAtCompany,dataset2.YearsSinceLastPromotion,dataset2.YearsWithCurrManager)
print(stats,p)
```

### H0: There is no significant differences in the YearsAtCompany and YearsSinceLastPromotion and YearsWithCurrManage
### Ha: There is significant differences in the YearsAtCompany and YearsSinceLastPromotion and YearsWithCurrManage
### The p value is 0.0 ,So p value is less than 0.5
### Hence Ho (null hypothesis) is rejected and Ha (alternative hypothesis) accepted.
### There is significant differences in the YearsAtCompany and YearsSinceLastPromotion and YearsWithCurrManage

# Mann-Whiteney Test
### Attrition and  DistanceFromHome

```python
from scipy.stats import mannwhitneyu
stats,p=mannwhitneyu(dataset2.Attrition,dataset2.DistanceFromHome)
print(stats,p)
```

# Kruskal-Wallis Test
### PercentSalaryHike	and StandardHours	and StockOptionLevel

```python
from scipy.stats import kruskal

stats,p = kruskal(dataset2.PercentSalaryHike,dataset2.StandardHours,dataset2.StockOptionLevel)
print(stats,p)
```

### H0: There is no significant differences in the PercentSalaryHike and StandardHours and StockOptionLevel
### Ha: There is significant differences in the PercentSalaryHike and StandardHours and StockOptionLevel
### The p value is 0.00 ,So p value is less than 0.5

### There is significant differences in the PercentSalaryHike and StandardHours and StockOptionLevel

# Chai-Square Test
### Attrition and Gender

```python
from scipy.stats import chi2_contingency

chitable = pd.crosstab(dataset2.Attrition,dataset2.Gender)
chitable
```

```python
stats,p,dof,expected=chi2_contingency(chitable)
print(stats,p)
```

### H0: There is no significant differences in the attrition yes and no for Gender
### Ha: There is significant differences in the attrition yes and no for Gender
### The p value is 0.24529482862926827 ,So p value is greater than 0.5
### Hence Ho (null hypothesis) is accepted and Ha (alternative hypothesis) rejected.
### There is no significant differences in the attrition yes and no for Gender.
